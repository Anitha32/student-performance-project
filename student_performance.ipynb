**Import Libraries**




```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load the dataset**

```python
df=pd.read_csv("/content/StudentsPerformance.csv")
df.head(5)
```

**Explore the Dataset**

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

**Rename Columns**

```python
df.columns=df.columns.str.replace(' ','_').str.lower()
df.head()
```

**Descriptive Statistics**

```python
df.describe()
```

**Grouped Analysis**

```python
df.groupby('gender')[['math_score','reading_score','writing_score']].mean()
```

```python
df.groupby('parental_level_of_education')[['math_score','reading_score','writing_score']].mean()
```

```python
df.groupby('test_preparation_course')[['math_score','reading_score','writing_score']].mean()
```

**Visualization**

```python
sns.boxplot(x='gender',y='math_score',data=df)
```

```python
sns.histplot(df['reading_score'], kde=True)
plt.title("Reading Score Distribution")
```

**Find Top performers**

```python
top_students = df[(df['math_score'] > 90) &
                  (df['reading_score'] > 90) &
                  (df['writing_score'] > 90)]
top_students
```

**Conclusion**

```python
conclusion = """
 CONCLUSION:

1. Female students performed better in reading and writing scores.
2. Students who completed the test preparation course scored higher on average.
3. Parental education level shows a mild effect on student performance.
4. Most students scored between 6080 in all three subjects.
5. A few students scored above 90 in all subjects  these can be considered top performers.

This analysis can help schools focus more on reading/writing support and encourage test preparation programs.
"""
print(conclusion)
```
